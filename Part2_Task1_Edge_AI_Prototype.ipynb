```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import os
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
import seaborn as sns
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D, Dropout
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam
import json
import time

# Set random seeds for reproducibility
tf.random.set_seed(42)
np.random.seed(42)

class RecyclableClassifier:
    def __init__(self, input_shape=(224, 224, 3), num_classes=6):
        """
        Initialize the recyclable item classifier

        Args:
            input_shape: Input image dimensions
            num_classes: Number of recyclable categories
        """
        self.input_shape = input_shape
        self.num_classes = num_classes
        self.model = None
        self.tflite_model = None
        self.class_names = ['cardboard', 'glass', 'metal', 'paper', 'plastic', 'trash']

    def create_lightweight_model(self):
        """
        Create a lightweight CNN model using MobileNetV2 as base
        MobileNetV2 is designed for mobile/edge devices
        """
        print("Creating lightweight model for edge deployment...")

        # Use MobileNetV2 as base model (pre-trained on ImageNet)
        base_model = MobileNetV2(
            input_shape=self.input_shape,
            include_top=False,  # Remove classification layer
            weights='imagenet'
        )

        # Freeze base model layers (transfer learning)
        base_model.trainable = False

        # Add custom classification layers
        inputs = tf.keras.Input(shape=self.input_shape)
        x = base_model(inputs, training=False)
        x = GlobalAveragePooling2D()(x)
        x = Dropout(0.2)(x)
        outputs = Dense(self.num_classes, activation='softmax')(x)

        self.model = Model(inputs, outputs)

        # Compile model
        self.model.compile(
            optimizer=Adam(learning_rate=0.0001),
            loss='categorical_crossentropy',
            metrics=['accuracy']
        )

        # Print model summary
        print("\nModel Architecture:")
        self.model.summary()

        return self.model

    def create_sample_data(self, num_samples=1000):
        """
        Create synthetic dataset for demonstration
        In real scenario, you'd load actual images
        """
        print(f"Creating sample dataset with {num_samples} images...")

        # Generate synthetic image data
        X = np.random.rand(num_samples, *self.input_shape)

        # Create labels (balanced distribution)
        y = np.random.randint(0, self.num_classes, num_samples)

        # Convert to categorical
        y_categorical = tf.keras.utils.to_categorical(y, self.num_classes)

        return X, y_categorical, y

    def train_model(self, X, y, validation_split=0.2, epochs=10, batch_size=32):
        """
        Train the lightweight model
        """
        print("Training the model...")

        # Data augmentation for better generalization
        datagen = ImageDataGenerator(
            rotation_range=20,
            width_shift_range=0.2,
            height_shift_range=0.2,
            horizontal_flip=True,
            zoom_range=0.2,
            validation_split=validation_split
        )

        # Create training and validation generators
        train_generator = datagen.flow(
            X, y,
            batch_size=batch_size,
            subset='training'
        )

        validation_generator = datagen.flow(
            X, y,
            batch_size=batch_size,
            subset='validation'
        )

        # Define callbacks
        callbacks = [
            tf.keras.callbacks.EarlyStopping(
                monitor='val_accuracy',
                patience=5,
                restore_best_weights=True
            ),
            tf.keras.callbacks.ReduceLROnPlateau(
                monitor='val_loss',
                factor=0.2,
                patience=3,
                min_lr=0.0001
            )
        ]

        # Train the model
        history = self.model.fit(
            train_generator,
            epochs=epochs,
            validation_data=validation_generator,
            callbacks=callbacks,
            verbose=1
        )

        return history

    def convert_to_tflite(self, optimize=True):
        """
        Convert trained model to TensorFlow Lite format for edge deployment
        """
        print("Converting model to TensorFlow Lite...")

        # Create TFLite converter
        converter = tf.lite.TFLiteConverter.from_keras_model(self.model)

        if optimize:
            # Apply optimizations for edge deployment
            converter.optimizations = [tf.lite.Optimize.DEFAULT]

            # Optional: Use representative dataset for quantization
            # This reduces model size and improves inference speed
            def representative_data_gen():
                for _ in range(100):
                    yield [np.random.rand(1, *self.input_shape).astype(np.float32)]

            converter.representative_dataset = representative_data_gen
            converter.target_spec.supported_ops = [tf.lite.OpsSet.TFLITE_BUILTINS_INT8]
            converter.inference_input_type = tf.uint8
            converter.inference_output_type = tf.uint8

        # Convert model
        self.tflite_model = converter.convert()

        # Save TFLite model
        with open('recyclable_classifier.tflite', 'wb') as f:
            f.write(self.tflite_model)

        print("TensorFlow Lite model saved as 'recyclable_classifier.tflite'")

        return self.tflite_model

    def evaluate_models(self, X_test, y_test):
        """
        Compare original model vs TFLite model performance
        """
        print("Evaluating models...")

        # Evaluate original model
        start_time = time.time()
        original_predictions = self.model.predict(X_test)
        original_inference_time = time.time() - start_time

        # Evaluate TFLite model
        interpreter = tf.lite.Interpreter(model_content=self.tflite_model)
        interpreter.allocate_tensors()

        input_details = interpreter.get_input_details()
        output_details = interpreter.get_output_details()

        start_time = time.time()
        tflite_predictions = []

        for i in range(len(X_test)):
            # Prepare input
            input_data = X_test[i:i+1].astype(np.float32)
            if input_details[0]['dtype'] == np.uint8:
                input_data = (input_data * 255).astype(np.uint8)

            interpreter.set_tensor(input_details[0]['index'], input_data)
            interpreter.invoke()

            output_data = interpreter.get_tensor(output_details[0]['index'])
            tflite_predictions.append(output_data[0])

        tflite_inference_time = time.time() - start_time
        tflite_predictions = np.array(tflite_predictions)

        # Calculate accuracies
        original_accuracy = np.mean(
            np.argmax(original_predictions, axis=1) == np.argmax(y_test, axis=1)
        )
        tflite_accuracy = np.mean(
            np.argmax(tflite_predictions, axis=1) == np.argmax(y_test, axis=1)
        )

        # Model sizes
        original_size = len(tf.keras.utils.serialize_keras_object(self.model))
        tflite_size = len(self.tflite_model)

        # Print results
        print("\n" + "="*50)
        print("MODEL COMPARISON RESULTS")
        print("="*50)
        print(f"Original Model:")
        print(f"  - Accuracy: {original_accuracy:.4f}")
        print(f"  - Inference Time: {original_inference_time:.4f}s")
        print(f"  - Model Size: ~{original_size/1024:.2f} KB")
        print(f"\nTensorFlow Lite Model:")
        print(f"  - Accuracy: {tflite_accuracy:.4f}")
        print(f"  - Inference Time: {tflite_inference_time:.4f}s")
        print(f"  - Model Size: {tflite_size/1024:.2f} KB")
        print(f"\nOptimization Results:")
        print(f"  - Size Reduction: {((original_size - tflite_size) / original_size * 100):.1f}%")
        print(f"  - Speed Improvement: {(original_inference_time / tflite_inference_time):.2f}x")

        return {
            'original_accuracy': original_accuracy,
            'tflite_accuracy': tflite_accuracy,
            'original_inference_time': original_inference_time,
            'tflite_inference_time': tflite_inference_time,
            'original_size': original_size,
            'tflite_size': tflite_size
        }

    def plot_training_history(self, history):
        """
        Plot training history
        """
        fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))

        # Plot accuracy
        ax1.plot(history.history['accuracy'], label='Training Accuracy')
        ax1.plot(history.history['val_accuracy'], label='Validation Accuracy')
        ax1.set_title('Model Accuracy')
        ax1.set_xlabel('Epoch')
        ax1.set_ylabel('Accuracy')
        ax1.legend()
        ax1.grid(True)

        # Plot loss
        ax2.plot(history.history['loss'], label='Training Loss')
        ax2.plot(history.history['val_loss'], label='Validation Loss')
        ax2.set_title('Model Loss')
        ax2.set_xlabel('Epoch')
        ax2.set_ylabel('Loss')
        ax2.legend()
        ax2.grid(True)

        plt.tight_layout()
        plt.show()

    def demonstrate_edge_inference(self):
        """
        Demonstrate real-time inference capabilities
        """
        print("\nDemonstrating Edge AI Inference...")

        # Load TFLite model
        interpreter = tf.lite.Interpreter(model_content=self.tflite_model)
        interpreter.allocate_tensors()

        input_details = interpreter.get_input_details()
        output_details = interpreter.get_output_details()

        # Simulate real-time inference
        for i in range(5):
            # Generate random test image
            test_image = np.random.rand(1, *self.input_shape).astype(np.float32)

            # Prepare input for TFLite
            if input_details[0]['dtype'] == np.uint8:
                test_image = (test_image * 255).astype(np.uint8)

            # Perform inference
            start_time = time.time()
            interpreter.set_tensor(input_details[0]['index'], test_image)
            interpreter.invoke()
            prediction = interpreter.get_tensor(output_details[0]['index'])
            inference_time = time.time() - start_time

            # Get predicted class
            predicted_class = np.argmax(prediction)
            confidence = np.max(prediction)

            print(f"Sample {i+1}: {self.class_names[predicted_class]} "
                  f"(Confidence: {confidence:.2f}, Time: {inference_time*1000:.2f}ms)")

def main():
    """
    Main execution function
    """
    print("Edge AI Recyclable Item Classifier")
    print("="*40)

    # Initialize classifier
    classifier = RecyclableClassifier()

    # Step 1: Create lightweight model
    model = classifier.create_lightweight_model()

    # Step 2: Create sample dataset
    X, y_categorical, y_labels = classifier.create_sample_data(num_samples=1000)

    # Split data
    X_train, X_test, y_train, y_test = train_test_split(
        X, y_categorical, test_size=0.2, random_state=42, stratify=y_labels
    )

    # Step 3: Train model
    history = classifier.train_model(X_train, y_train, epochs=5)  # Reduced epochs for demo

    # Step 4: Plot training history
    classifier.plot_training_history(history)

    # Step 5: Convert to TensorFlow Lite
    tflite_model = classifier.convert_to_tflite(optimize=True)

    # Step 6: Evaluate models
    results = classifier.evaluate_models(X_test, y_test)

    # Step 7: Demonstrate edge inference
    classifier.demonstrate_edge_inference()

    print("\n" + "="*50)
    print("EDGE AI BENEFITS SUMMARY")
    print("="*50)
    print("1. Reduced Latency: Real-time inference without network calls")
    print("2. Privacy: Data processing happens locally on device")
    print("3. Offline Operation: Works without internet connectivity")
    print("4. Reduced Bandwidth: No need to send data to cloud")
    print("5. Lower Costs: Reduced cloud computing expenses")
    print("6. Scalability: Distributed processing across edge devices")

    return classifier, results

if __name__ == "__main__":
    classifier, results = main()
```
